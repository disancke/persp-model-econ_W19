```python
%%latex
5.1

$$\max\limits_{w_2\in [0,W_1]} u(W_1-W_2)$$
```

```python
%%latex
5.2

The condition for the optimal amount to leave for period $W_3$ in period 2 is
$$\max\limits_{w_3\in [0,W_2]} u(W_2-W_3)$$

The condition for the optimal amount to leave for period $W_2$ in period 1 is
$$\max\limits_{w_2\in [0,W_1]} (u(W_1-W_2)+\beta \max\limits_{w_3\in [0,W_2]} u(W_2-W_3))$$
```

```python
%%latex
5.3

For period $W_2$ in period 1, we have
$$\max\limits_{w_2\in [0,W_1]} (u(W_1-W_2)+\beta \max\limits_{w_3\in [0,W_2]} u(W_2-W_3)+\beta^2 \max\limits_{w_4\in [0,W_3]} u(W_3-W_4)))$$

For period $W_3$ in period 2, we have
$$\beta \max\limits_{w_3\in [0,W_2]} u(W_2-W_3)+\beta^2 \max\limits_{w_4\in [0,W_3]} u(W_3-W_4)))$$

For period $W_4$ in period 3, we have
$$\beta^2 \max\limits_{w_4\in [0,W_3]} u(W_3-W_4)$$
```

```python
%%latex
5.4
We want to differentiate (5.7) and have
$$u'(W_{T-1}-W_T)+\beta u'(W_T)=0$$
Substituting $W_T=\psi_{T-1}(W_{T-1})$, we have
$$V_{T-1}(W_{T-1})=u(W_{T-1}-\psi_{T-1}(W_{T-1}))+\beta u(\psi_{T-1}(W_{T-1}))$$

```

```python
%%latex
5.5
When $u(x)=ln(x)$, from 5.4 we have $\psi_{T}(\bar{W_{T}})=0$, thus $\psi_{T-1}\neq\psi_{T}.$
Also from 5.4 we have $$V_{T-1}(W_{T-1})=ln(\frac{\bar{W}}{1+\beta})+\beta ln(\frac{\beta \bar{W}}{1+\beta})$$
Thus $V_T\neq V_{T-1}.$
```

```python
%%latex
5.6
The Bellman equation at time T-2 is
$$V_{T-2}(W_{T-2})=\max\limits_{W_{T-1}} ln(W_{T-2}-W_{T-1})+\beta ln(\frac{W_{T-1}}{1+\beta})+\beta^2 ln(\frac{\beta W_{T-1}}{1+\beta})$$

The FOC of the equation above is
$$\frac{\beta (1+\beta)}{\psi_{T-2}(W_{T-2})}=\frac{1}{W_{T-2}-\psi_{T-2}(W_{T-2})}$$

Thus we have the analytical solutions
$$\psi_{T-2}(W_{T-2})=\frac{\beta+\beta^2}{1+\beta+\beta^2}W_{T-2}$$
$$V_{T-2}(W_{T-2})=ln(\frac{W_{T-2}}{1+\beta+\beta^2})+\beta ln(\frac{\beta W_{T-2}}{1+\beta+\beta^2})+\beta^2 ln(\frac{\beta^2 W_{T-2}}{1+\beta+\beta^2})$$
```

```python
%%latex
5.7
The solutions are:
$$\psi_{T-s}(W_{T-s})=\frac{\sum\limits_{i=1}^{s}\beta^{i}}{\sum\limits_{i=0}^{s}\beta^{i}}W_{T-s}$$
$$V_{T-s}(W_{T-s}) = \sum\limits_{i=0}^{s}\beta^{i} · ln\left(\frac{\beta^{i}}{\sum\limits_{j=0}^{s}\beta^{j}} W_{T-s} \right)$$
```

```python
%%latex
5.8
The bellman equation under infinite horizon is:
$$V(W)=\max\limits_{W'\in [0,W]} u(W-W')+\beta V(W')$$
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from mpl_toolkits.mplot3d import Axes3D
```

```python
#5.9

W_vec=np.linspace(0.01, 1, 100)
```

```python
#5.10
beta=0.9
N=100

def utility(c):
    util = np.log(c)
    return util

V_init = utility(W_vec)
V_prime = np.zeros(N).reshape(N,1)

c_mat = np.tile(W_vec.reshape((N,1)),(1,N)) - np.tile(W_vec.reshape((1,N)),(N,1))
c_pos = c_mat > 0
c_mat[~c_pos] = 1e-7 #Replace non-zero values
u_mat = utility(c_mat)

V_new = (u_mat+beta * V_prime).max(axis=1)
index = np.argmax(u_mat+beta * V_prime, axis=1)

print(W_vec[index])
```

```python
plt.plot(V_new)
plt.title("Value function")
plt.show()
```

```python
plt.plot(W_vec,W_vec[index])
plt.title("Policy function")
plt.show()
```

```python
dist = ((V_new - V_init)**2).sum()
print(dist)
```

```python
#5.12

V_init = utility(W_vec)
V_prime = V_new

c_mat = np.tile(W_vec.reshape((N,1)),(1,N)) - np.tile(W_vec.reshape((1,N)),(N,1))
c_pos = c_mat > 0
c_mat[~c_pos] = 1e-7 #Replace non-zero values
u_mat = utility(c_mat)

V_prime=np.tile(V_new.reshape((1,N)),(N,1))
V_prime[~c_pos]=-1e+7

V_new1 = (u_mat+beta * V_prime).max(axis=1)
index = np.argmax(u_mat+beta * V_prime, axis=1)
```

```python
#5.16
sigma = 0.5
mu = sigma*4
M=7

epsilon = np.linspace(mu-3*sigma, mu+3*sigma, M)
Gamma = lambda x: norm(mu, sigma).pdf(x)
Gamma(epsilon)
```

```python
#5.17
V_init = utility(W_vec)
V_prime = np.zeros(N).reshape(N,1)

c_mat = np.tile(W_vec.reshape((N,1)),(1,N)) - np.tile(W_vec.reshape((1,N)),(N,1))
c_pos = c_mat > 0
c_mat[~c_pos] = 1e-7 #Replace non-zero values
u_mat = utility(c_mat)
cube = np.array([u_mat * i for i in epsilon])

V_init = np.zeros((N, M))
V1 = V_init @ Gamma(epsilon).reshape((M,1))
V1_mat = np.tile(V1.reshape((1,N)), (N,1))
V1_mat[~c_pos] = -9e+4
V1_cube = np.array([V1_mat for i in range(M)])

VT_cube = cube + beta * V1_cube
V_new = np.zeros((N,M))
W_new = np.zeros((N,M))

for i in range(N):
    VW_cube = VT_cube[:, i, :]
    V_new[i] = VW_cube.max(axis=1)
    index = np.argmax(VW_cube, axis=1)
    W_new[i] = W_vec[index]
```

```python
#5.18
dist = ((V_init - V_new)**2).sum()
print(dist)
```

```python
#5.19

V_init = V_new
V1 = V_init @ Gamma(epsilon).reshape((M,1))
V1_mat = np.tile(V1.reshape((1,N)), (N,1))
V1_mat[~c_pos] = -9e+4
V1_cube = np.array([V1_mat for i in range(M)])

VT_cube = cube + beta * V1_cube
V_new1 = np.zeros((N,M))
W_new1 = np.zeros((N,M))

for i in range(N):
    VW_cube = VT_cube[:, i, :]
    V_new1[i] = VW_cube.max(axis=1)
    index = np.argmax(VW_cube, axis=1)
    W_new1[i] = W_vec[index]
```

```python
maxiters = 500
toler = 1e-10
dist = 10.0
VF_iter = 0
V_init = np.zeros((N,M))
while dist > toler and VF_iter < maxiters:
    VF_iter += 1

    V1 = V_init @ Gamma(epsilon).reshape((M,1))
    V1_mat = np.tile(V1.reshape((1,N)), (N,1))
    V1_mat[~c_pos] = -9e+4
    V1_cube = np.array([V1_mat for i in range(M)])

    VT_cube = cube + beta * V1_cube
    V_new = np.zeros((N,M))
    W_new = np.zeros((N,M))

    for i in range(N):
        VW_cube = VT_cube[:, i, :]
        V_new[i] = VW_cube.max(axis=1)
        index = np.argmax(VW_cube, axis=1)
        W_new[i] = W_vec[index]
    
    dist=np.sum((V_init-V_new)**2)
    print('Iter=', VF_iter, ', distance =', dist)
    V_init = V_new

print('V(W)', V_init)
print("It takes", VF_iter, "iterations")
```
